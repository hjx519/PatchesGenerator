```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
output_data = np.load('patches/patch0.npy')
for i in range(output_data.shape[0]):
    plt.figure()  
    if i == 0:  
        plt.imshow(np.abs(output_data[i,:,:]), cmap='jet')
        plt.colorbar(label='s')
        plt.title('T2 map')
    else:  
        plt.imshow(np.abs(output_data[i,:,:]), cmap='gray')
        plt.colorbar(label='Grayscale value')
        plt.title(f'phase-cycled bssfp {i}')
    plt.show()
```

```python
plt.figure()
y = 5
x = 5
point = output_data[-6:,y, x]
plt.scatter(point.real, point.imag)
plt.title(f'Pixel (y={y}, x={x})')
```
